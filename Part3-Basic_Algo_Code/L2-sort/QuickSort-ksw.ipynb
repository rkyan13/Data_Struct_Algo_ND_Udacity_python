```python
def quicksort_ascending(input_list):
    iplist_len = len(input_list) 
    
    if(iplist_len==0 or iplist_len ==1):
        sorted_list = input_list    
        return sorted_list
    
    elif(iplist_len ==2):
        if(input_list[0]>input_list[1]):
            temp          = input_list[0]
            input_list[0] = input_list[1]
            input_list[1] = temp
            sorted_list   = input_list
        else:
            sorted_list = input_list
            
        return sorted_list
    
    else:
        last_idx                       = iplist_len-1
        pivot_index, rough_sorted_list = rough_sort_ascending(input_list)
        pivot                          = rough_sorted_list[pivot_index]
        if  (pivot_index == 0):
            quick_sorted_tail  = quicksort_ascending(rough_sorted_list[1:last_idx])
            sorted_list        = [pivot]+quick_sorted_tail
        elif(pivot_index == last_idx):
            quick_sorted_head  = quicksort_ascending(rough_sorted_list[0:last_idx-1])
            sorted_list        = quick_sorted_head+[pivot]
        else:
            quick_sorted_head  = quicksort_ascending(rough_sorted_list[0:pivot_index-1])
            quick_sorted_tail  = quicksort_ascending(rough_sorted_list[pivot_index+1:last_idx])
            sorted_list        = quick_sorted_head+[pivot]+quick_sorted_tail
            
    
    
    return sorted_list

'''
rough_sort 
--> puts the pivot in the correct position
NOTE:pivot need not necessarily be at the center of the list

--> Roughly groups elements lower than the pivot to the left of the pivot. 
They are smaller than the pivot,but aren't sorted(in ascending order).They could be in any order amongst themselves
Example: If the pivot is 10. 
         To its left could be [7,3,0,-5]. They aren't sorted, but all are smaller than 10

-->Roughly groups elements higher than the pivot to the right of the pivot.
They are larger than the pivot,but aren't sorted(in ascending order).They could be in any order amongst themselves
Example: If the pivot is 10. 
         To its right could be [125,2500,23,99,75,56,19,550].They aren't sorted,but are all larger than 10

rough_sort output looks like 
[7,3,0,-5, 10,125,2500,23,99,75,56,19,550]



find_first_larger_from_left
--> helper function to find the first element larger than the pivot as you traverse the list from left to right


find_first_smaller_from_right
--> helper function to find the first element smaller than the pivot as you traverse the list from right to left

'''



def rough_sort_ascending(input_list):
    pivot               = input_list[-1]
    pivot_index         = len(input_list)-1
    rough_sort_complete = False
    
    while(rough_sort_complete == False):
        first_larger_from_left_idx   = find_first_larger_from_left(input_list)
        first_smaller_from_right_idx = find_first_smaller_from_right(input_list)
        
        #if first_larger_from_left_idx<first_smaller_from_right_idx swap the two'''
        if(first_larger_from_left_idx<first_smaller_from_right_idx):
            temp                                     = input_list[first_larger_from_left_idx]
            input_list[first_larger_from_left_idx]   = input_list[first_smaller_from_right_idx]
            input_list[first_smaller_from_right_idx] = temp
            
        #if first_larger_from_left_idx>first_smaller_from_right_idx this means rough sort is complete.
        #Just the put the pivot in the correct position and return
        #You only need to swap if the pivot_idx is different from first_larger_frm_left_idx
        #Ex. input_list = [1,2,3,4,5], pivot = 5,No elements larger than 5.
        #Pivot in the right place already.No need to swap
        else:
            rough_sort_complete = True
            if(pivot_index !=first_larger_from_left_idx):
                pivot_index         = first_larger_from_left_idx
                temp                                     = input_list[first_larger_from_left_idx]
                input_list[first_larger_from_left_idx]   = input_list[pivot_index]
                input_list[pivot_index]                  = temp
                
    
    return pivot_index, input_list



'''
corner case 
for input_list = [1,2,3,4,5], pivot = 5
there are no elements larger than 5, 
to take care of this corner case set first_larger_frm_left_idx = pivot_idx @start of for loop
'''
def find_first_larger_from_left(input_list):
    pivot     = input_list[-1]
    pivot_idx = len(input_list)-1
    first_larger_frm_left_idx = pivot_idx
    for idx, number in enumerate(input_list):
        if(number>pivot):
            first_larger_frm_left_idx = idx
            break
    
    return first_larger_frm_left_idx


'''
corner case 
for input_list = [5,4,3,2,1], pivot = 1
there are no elements smaller than 1, 
to take care of this corner case set first_smaller_frm_right_idx = 0 @start of while loop
'''
def find_first_smaller_from_right(input_list):
    pivot                      = input_list[-1]
    pivot_idx                  = len(input_list)-1
    first_smaller_frm_right_idx = 0
    idx                        = pivot_idx -1
    while(idx>=0):
        if(input_list[idx]<pivot):
            first_smaller_frm_right_idx = idx
            break
        idx-=1
    
    return first_smaller_frm_right_idx
            
        
```
