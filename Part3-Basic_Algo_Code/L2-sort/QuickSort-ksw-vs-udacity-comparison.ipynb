
### rough_sort   
i) puts the pivot in the correct position  
NOTE:pivot need not necessarily be at the center of the list  

ii) Roughly groups elements lower than the pivot to the left of the pivot.   
They are smaller than the pivot,but aren't sorted(in ascending order).They could be in any order amongst themselves  
Example: If the pivot is 10.   
         To its left could be [7,3,0,-5]. They aren't sorted, but all are smaller than 10  

iii) Roughly groups elements higher than the pivot to the right of the pivot.  
They are larger than the pivot,but aren't sorted(in ascending order).They could be in any order amongst themselves  Example: If the pivot is 10.   
         To its right could be [125,2500,23,99,75,56,19,550].They aren't sorted,but are all larger than 10  

rough_sort output looks like  
[7,3,0,-5, 10,125,2500,23,99,75,56,19,550]  



### find_first_larger_from_left
helper function to find the first element larger than the pivot as you traverse the list from left to right  
corner case:  
for number_list = [1,2,3,4,5], pivot = 5, there are no elements larger than 5   
to take care of this corner case set first_larger_frm_left_idx = pivot_idx @start of while loop  


### find_first_smaller_from_right
helper function to find the first element smaller than the pivot as you traverse the list from right to left    
corner case:  
for number_list = [5,4,3,2,1], pivot = 1,there are no elements smaller than 1  
to take care of this corner case set first_smaller_frm_right_idx = 0 @start of while loop



```python
DEBUG = 0

def quicksort_ascending(number_list, start_idx=0, end_idx=None):

    if(end_idx == None):
            list_len = len(number_list) 
            end_idx  = list_len-1
    else:
        list_len = end_idx-start_idx +1

    if(list_len==0 or list_len ==1):
        return 
    
    elif(list_len ==2):
        if(number_list[start_idx]>number_list[end_idx]):
            temp                   = number_list[start_idx]
            number_list[start_idx] = number_list[end_idx]
            number_list[end_idx]   = temp
          
        return 
    
    else:            
        pivot_index = rough_sort_ascending(number_list,start_idx,end_idx)
        pivot       = number_list[pivot_index]
        if  (pivot_index == start_idx):
            #Quicksort the tail
            quicksort_ascending(number_list, start_idx+1, end_idx)
        elif(pivot_index == end_idx):
            #Quicksort the head
            quicksort_ascending(number_list, start_idx, end_idx-1)
        else:
            #Quicksort the head
            quicksort_ascending(number_list,start_idx,pivot_index-1)
            #Quicksort the tail
            quicksort_ascending(number_list,pivot_index+1,end_idx)   
        
    
    return



def rough_sort_ascending(number_list,start_idx,end_idx):
    
    pivot_index  = end_idx
    pivot        = number_list[pivot_index]
    rough_sort_complete = False
    
    while(rough_sort_complete == False):
        first_larger_from_left_idx   = find_first_larger_from_left(number_list,start_idx,end_idx)
        first_smaller_from_right_idx = find_first_smaller_from_right(number_list,start_idx,end_idx)
        
        #if first_larger_from_left_idx<first_smaller_from_right_idx swap the two
        if(first_larger_from_left_idx<first_smaller_from_right_idx):
            temp                                      = number_list[first_larger_from_left_idx]
            number_list[first_larger_from_left_idx]   = number_list[first_smaller_from_right_idx]
            number_list[first_smaller_from_right_idx] = temp
            
        #if first_larger_from_left_idx>first_smaller_from_right_idx this means rough sort is complete.
        #Just the put the pivot in the correct position and return
        else:
            rough_sort_complete = True
            if(pivot_index !=first_larger_from_left_idx):
                temp                                     = number_list[first_larger_from_left_idx]
                number_list[first_larger_from_left_idx]  = number_list[pivot_index]
                number_list[pivot_index]                 = temp
                pivot_index                              = first_larger_from_left_idx               
    
    return pivot_index




def find_first_larger_from_left(number_list,start_idx,end_idx):
    pivot_idx                 = end_idx
    pivot                     = number_list[pivot_idx]
    first_larger_frm_left_idx = pivot_idx
    
    idx                       = start_idx
    while(idx<=end_idx):
        if(number_list[idx]>pivot):
            first_larger_frm_left_idx = idx
            break
        idx+=1
        
    return first_larger_frm_left_idx



def find_first_smaller_from_right(number_list,start_idx,end_idx):
    pivot_idx    = end_idx
    pivot        = number_list[pivot_idx]
    first_smaller_frm_right_idx = start_idx
    
    idx                         = pivot_idx -1
    while(idx>=start_idx):
        if(number_list[idx]<pivot):
            first_smaller_frm_right_idx = idx
            break
        idx-=1
    
    return first_smaller_frm_right_idx
            
print(".......")        
```

```python
def quicksort(items):
    sort_all(items, 0, len(items) - 1)
    
def sort_all(items, begin_index, end_index):
    if end_index <= begin_index:
        return
    
    pivot_index = sort_a_little_bit(items, begin_index, end_index)
    sort_all(items, begin_index, pivot_index - 1)
    sort_all(items, pivot_index + 1, end_index)
    
def sort_a_little_bit(items, begin_index, end_index):    
    left_index = begin_index
    pivot_index = end_index
    pivot_value = items[pivot_index]

    while (pivot_index != left_index):

        item = items[left_index]

        if item <= pivot_value:
            left_index += 1
            continue

        items[left_index] = items[pivot_index - 1]
        items[pivot_index - 1] = pivot_value
        items[pivot_index] = item
        pivot_index -= 1
    
    return pivot_index
```
