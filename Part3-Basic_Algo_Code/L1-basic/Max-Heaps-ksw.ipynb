```python
import math

class Max_Heap(): 
    def __init__(self, arr = None):
        self.arr = arr
        if(arr == None):
            self.last_idx   = 0
        else:
            heap_sz  = 0
            for item in enumerate(self.arr):
                if (item !=None):
                    heap_sz+=1
                else:
                    break
            self.last_idx = heap_sz -1

    
    def sizeof_arr(self):
        if(self.arr == None):
            arr_sz= 0
        else:
            arr_sz = len(self.arr)
        
        return arr_sz
    
    def sizeof_heap(self):
        heap_sz  = 0
        for item in self.arr:
            if (item !=None):
                heap_sz+=1
            else:
                break
        
        return heap_sz
    
        
    def is_full(self):
        if (self.last_idx== len(self.arr) -1):
            return True
        else:
            return False
        
    def max_heapify(self, debug =0):
        heap_size = self.sizeof_heap()
        tree_ht   =  math.ceil(math.log(heap_size+1,2))
        if(debug == 1):
            print("tree_ht =", tree_ht)
        
        #tdl: top down level
        #the root of the cbt(complete binary tree) is tdl = 0.
        #the root's children are at tdl = 1, root's grandchildren at tdl = 2 and so on,
        #bottom most level which has only leaves    :tdl = tree_ht-1: this level has zero swaps. 
        #So start with Penultimate level from bottom:tdl = tree_ht-2: this level has 1 swap per element in this level
        TDL       = tree_ht -2
        TDL_swaps = 1
        
        #LOOP1: iterate over all TDLs---------------------------------------------
        while(TDL >=0):
            if(debug == 1):
                print("\n-----------------TDL=",TDL,":TDL_swaps=", TDL_swaps,"-----------------")
                
            TDL_num_elements = pow(2,TDL)
            TDL_start_idx    = TDL_num_elements - 1
            TDL_end_idx      = TDL_start_idx*2
            
            #--------LOOP2: interate over all ELEMENTS in a TDL-------------------
            for idx in range(TDL_start_idx, TDL_end_idx+1):
                parent_idx = idx
                self._max_down_swap(TDL,TDL_swaps,parent_idx, debug)
            #--------EOF: for LOOP2-----------------------------------------------
            
            if(debug == 1):
                print("EOF TDL: heap(array) =",self.arr)
                print("-------------------------------------------------------")                
            TDL       -=1
            TDL_swaps +=1
        #--------EOF: while LOOP1------------------------------------------------
        
        if(debug == 1):
            print("\nFINAL:Post max_heapify:",self.arr)  
        
        return
            
    #_max_down_swap: helper function for max_heapify
    def _max_down_swap(self,TDL,TDL_swaps,parent_idx,debug = 0):        
        #LOOP3: complete all the TDL_swaps
        heap_size = self.sizeof_heap()
        curr_TDL        = TDL
        curr_parent_idx = parent_idx 
        for curr_TDL_swap in range(1,TDL_swaps+1):
            if(debug == 1):
                print("TDL=",TDL, ":TDL_swaps=", TDL_swaps,":curr_TDL_swap=", curr_TDL_swap, \
                      ":curr_TDL=", curr_TDL, ":curr_parent_idx=",curr_parent_idx, ":curr_parent = ",self.arr[curr_parent_idx])
                
            left_idx  = 2*curr_parent_idx +1
            right_idx = 2*curr_parent_idx +2
            
            if  (left_idx>=heap_size): #If neither left,nor right child exists
                break            
            elif(right_idx>=heap_size):#If only left child exists
                greater_child_idx = left_idx
            else:                      #If both left and right child exist
                if(self.arr[left_idx]>self.arr[right_idx]):
                    greater_child_idx = left_idx
                else:
                    greater_child_idx = right_idx
            
            if(self.arr[curr_parent_idx]<self.arr[greater_child_idx]): #child is greater
                temp                        = self.arr[curr_parent_idx]
                self.arr[curr_parent_idx]   = self.arr[greater_child_idx]
                self.arr[greater_child_idx] = temp
                curr_parent_idx             = greater_child_idx
                curr_TDL                    += 1
            else:#parent is greater(already a local max-heap).No need to swap
                break
        
        return
    
    def insert_to_max_heap(self, value, debug = 0):
        # if array is full double the capacity of the array
        if(self.is_full() == True):
            temp_arr = self.arr
            self.arr = [None for _ in range(0, 2*len(temp_arr))]
            for idx in range(0,len(temp_arr)):
                self.arr[idx] = temp_arr[idx]

        insert_idx = self.last_idx+1
        self.arr[insert_idx] = value
        self.last_idx +=1
        self._max_up_heapify()
        
    #_max_up_heapify: helper function for insert_to_max_heap
    def _max_up_heapify(self,debug = 0):        
        heap_size      = self.sizeof_heap()
        child_idx      = heap_size -1
        parent_idx     = int(0.5*(child_idx-1))
        
        if(parent_idx >=0):
            heap_traversed = False
        else:
            heap_traversed = True
        
        while(heap_traversed == False):
            if(self.arr[child_idx]>self.arr[parent_idx]): #child is greater
                temp                 = self.arr[parent_idx]
                self.arr[parent_idx] = self.arr[child_idx]
                self.arr[child_idx]  = temp
                #Update child-idx
                child_idx            = parent_idx
                #calculate new parent
                parent_idx           = int(0.5*(child_idx-1))
                if(parent_idx <0):
                    heap_traversed = True                
            else:#parent is greater(already a local max-heap).No need to swap
                heap_traversed = True
                
        return
    
    def max_heap_delete_root(self, heap_size=None,debug = 0):
        remove_idx_found,remove_idx, value_removed = self.remove_from_max_heap(value_to_remove="root", \
                                                                               heap_size=heap_size,\
                                                                               debug = debug)
        return remove_idx_found,remove_idx, value_removed
    
    
    def remove_from_max_heap(self, value_to_remove="root", heap_size=None,debug = 0):
        if(heap_size == None):
            heap_size       = self.sizeof_heap()
        if(heap_size == 0):
            remove_idx_found = False
            remove_idx       = None
            value_to_remove  = None
        else:
            #If you want the root of the heap removed
            if(value_to_remove == "root"):
                remove_idx_found = True
                remove_idx       = 0 #i.e remove the root
                value_to_remove  = self.arr[0]
            #If you want some other element removed
            else: 
                remove_idx_found = False
                remove_idx       = None
                for i in range(heap_size):
                    if(self.arr[i]==value_to_remove):
                        remove_idx_found = True
                        remove_idx = i
                        value_to_remove  = self.arr[remove_idx]
                        break

            if(remove_idx_found == True):
                self.arr[remove_idx]   = self.arr[heap_size -1]
                self.arr[heap_size -1] = None
                heap_size = heap_size -1
                self._max_down_heapify(parent_idx = remove_idx,\
                                       heap_size  = heap_size, \
                                       debug      = debug)
        
        return remove_idx_found,remove_idx, value_to_remove
        
    
    #_max_down_heapify: helper function for remove_from_max_heap
    #default: parent_idx = 0 i.e. root
    def _max_down_heapify(self,parent_idx= 0, heap_size=None, debug = 0):  
        if(heap_size == None):
            heap_size       = self.sizeof_heap()
        curr_parent_idx = parent_idx #if parent_idx = 0 , will start with root
        heap_traversed  = False
   
        while(heap_traversed == False):
            left_idx  = 2*curr_parent_idx +1
            right_idx = 2*curr_parent_idx +2
            if  (left_idx>=heap_size): #If neither left,nor right child exists
                heap_traversed = True
                break            
            elif(right_idx>=heap_size):#If only left child exists
                greater_child_idx = left_idx
            else:                      #If both left and right child exist
                if(self.arr[left_idx]>self.arr[right_idx]):
                    greater_child_idx = left_idx
                else:
                    greater_child_idx = right_idx

            if(self.arr[curr_parent_idx]<self.arr[greater_child_idx]): #child is greater
                temp                        = self.arr[curr_parent_idx]
                self.arr[curr_parent_idx]   = self.arr[greater_child_idx]
                self.arr[greater_child_idx] = temp
                curr_parent_idx             = greater_child_idx
            else:#parent is greater(already a local max-heap).No need to swap
                heap_traversed = True
                break
                
        return
    
    
    def ascending_heap_sort(self,debug =0):
        #make the input array into a valid max heap
        self.max_heapify()
        
        #Now begin the sorting process using delete_root method
        arr_size  = self.sizeof_arr()
        heap_size = arr_size
        while(heap_size !=0):
            _,_, value_removed = self.max_heap_delete_root(heap_size)
            self.arr[heap_size-1] = value_removed
            heap_size = heap_size -1

        
    
    def descending_heap_sort(self,debug =0):
        self.ascending_heap_sort()
        array_size      = self.sizeof_arr()
        array_half_size = int(array_size/2)
        for i in range(array_half_size):
            temp                     = self.arr[i]
            self.arr[i]              = self.arr[array_size-i-1]
            self.arr[array_size-i-1] = temp

            
        
    
    
    
```

## Unit Tests: _max_up_heapify 
_max_up_heapify will typically not be used by itself. It will be used in the insert function.  
With the exception of the newly inserted element, the CBT is already a max_heap.  
Hence_ max_up_heapify only propagates the newly inserted element upwards to the correct height/ level to restores the heap order property.It does not check the heap order property of the entire CBT


```python
input_array = [30, 20, 25, 15, 5, 0, 10, 40, None, None, None, None, None, None, None]
my_heap = Max_Heap(input_array)
my_heap._max_up_heapify(debug = 1)
print("Insert 40:",my_heap.arr)

input_array = [30, 20, 25, 15, 5, 0, 10, 21, None, None, None, None, None, None, None]
my_heap = Max_Heap(input_array)
my_heap._max_up_heapify(debug = 1)
print("Insert 21:",my_heap.arr)

input_array = [30, 20, 25, 15, 5, 0, 10, 26, None, None, None, None, None, None, None]
my_heap = Max_Heap(input_array)
my_heap._max_up_heapify(debug = 1)
print("Insert 26:",my_heap.arr)

input_array = [30, 20, 25, 15, 5, 0, 10, 17, None, None, None, None, None, None, None]
my_heap = Max_Heap(input_array)
my_heap._max_up_heapify(debug = 1)
print("Insert 17:",my_heap.arr)

input_array = [30, 20, 25, 15, 5, 0, 10, 9, None, None, None, None, None, None, None]
my_heap = Max_Heap(input_array)
my_heap._max_up_heapify(debug = 1)
print("Insert  9:",my_heap.arr)

input_array = [30, 20, 25, 15, 5, 0, 10, -1, None, None, None, None, None, None, None]
my_heap = Max_Heap(input_array)
my_heap._max_up_heapify(debug = 1)
print("Insert -1:",my_heap.arr)
```
