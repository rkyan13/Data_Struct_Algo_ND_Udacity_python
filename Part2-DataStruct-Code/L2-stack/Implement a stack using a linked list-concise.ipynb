```python
class Node():
    def __init__(self,value=None):
        self.value = value
        self.next  = None
     
    def set_value(self,value):
        self.value = value

    def get_value(self):
        return self.value
    
    def set_next(self, next_node):
        self.next = next_node
        
    def print_content(self):
        print("value =", self.value)
        print("next  =", self.next)

#Exercise : Implement a linked list(instructions: it should contain a head. It cannot contain a tail or number of elements)
class LinkedList():
    def __init__(self):
        self.head = Node()
        
    #prepend: add element at the head
    def prepend(self,value):
        # if the Linked List has been initalised with a head but no value in the head yet
        if(self.head.value == None):
            self.head.value = value
            return
        else:
            node_to_append = Node(value)
            node_to_append.set_next(self.head)
            self.head = node_to_append
            return

    #append: add element at the tail
    #Equivalent of "push" on a stack
    def append(self, value):
        # if the Linked List has been initalised with a head but no value in the head yet
        if(self.head.value == None):
            self.head.value = value
            return
        else:
            ptr = self.head
            while(ptr.next !=None):
                ptr = ptr.next
            
            node_to_append = Node(value)
            ptr.next       = node_to_append
            return
    
    #Delete @head : delete the first element and move the head 1 place forward
    #Also return the value that was deleted
    def delete_at_head(self):
        #stack is empty already
        if(self.head.value == None):
            return None
        #head is the only Node
        elif(self.head.next ==None):
            popped_value       = self.head
            self.head.value    = None
            return popped_value
        #there are atleast 2 nodes (1st head and 1 other). Move the head to the second node
        else: 
            popped_value    = self.head.value
            self.head.value = self.head.next.value 
            return popped_value
        
        return 
        
    
    #Delete @tail : delete the last element and move the tail 1 place back
    #Also return the value that was deleted
    #Equivalent of "pop" on a stack
    def delete_at_tail(self):
        #stack is empty already
        if(self.head.value == None):
            return None
        #head is the only node
        elif(self.head.next == None):
            popped_value = self.head.value
            self.head.value = None
            return popped_value
        
        #there are atleast 2 nodes (1 head and 1 other)
        #start with: Previous Node-Head, Current Node- 2nd node, Next Node- whatever 2nd points to (this could be None)
        #Keep traversing the list until you encounter the tail(i.e when curr_ptr.next is None , tail is encountered )
        else: 
            prev_ptr = self.head
            curr_ptr = self.head.next
            while(curr_ptr.next != None):
                prev_ptr = curr_ptr
                curr_ptr = curr_ptr.next
            popped_value = curr_ptr.value
            prev_ptr.next = None 
            return popped_value
        
        return   

        
    #Delete first occurence of a value
    def delete(self,value):
        #if head is a match
        if(self.head.value == value):
            match_found = 1
            curr_loc    = 0
            #If the head was the only node in the list, set the value to None
            if(self.head.next == None):
                self.head.value = None
            #Else move the head to the next position (but the previous head node will just be dangling ??)
            else:
                self.head = self.head.next
                
        #head is not a match
        else:
            prev_ptr    = self.head
            curr_ptr    = self.head.next
            match_found = 0
            curr_loc    = 1
            while(curr_ptr !=None):
                if(curr_ptr.value == value):
                    prev_ptr.next = curr_ptr.next
                    match_found = 1
                    break
                else:
                    prev_ptr = curr_ptr
                    curr_ptr = curr_ptr.next
                    curr_loc +=1            
                    
        if(match_found == 1):
            print("match_found to delete @loc =", curr_loc)
        else:
            print("match not found")
        
        return
    
    def print_contents(self):
        curr = self.head
        while(curr!=None):
            print(curr.value)
            curr = curr.next
        
        return

print("...")
```

```python
class Stack(LinkedList):
    def __init__(self):
        LinkedList.__init__(self)

    
    def push(self, value):
        self.append(value)

    
    def pop(self):
        return self.delete_at_tail()

    
    def size(self):
        count = 0
        if (self.head.value == None):
            return count
        else:
            curr =self.head
            while(curr!=None ):
                count +=1
                curr = curr.next        
            return count

    
    def is_empty(self):
        if(self.size()==0): 
            return True
        else:
            return False
        
    def print_stack(self):
        self.print_contents()

    


print(".....")
```
