```python
from collections import deque

class Queue:
    def __init__(self,values = None):
        self.queue = deque()
        if(values):
            for i in values:
                self.queue.append(i)
    
    #Think of the deque as a list from left to right
    #I am going to enq on the right and deq on the left (this is the opposite of the udacity videos)
    def enq(self, value):
        self.queue.append(value)
    
    def deq(self):
        return self.queue.popleft()
    
    def peek_head(self):
        return self.queue[0]
    
    def peek_tail(self):
        return self.queue[-1]
    
    def qsize(self):
        return len(self.queue)
    
    def is_empty(self):
        if(len(self.queue)==0):
            return True
        else:
            return False
    
    def print(self):
        print(self.queue)
        
    

my_q = Queue()
print("size =",my_q.qsize())
my_q.print()
my_q.enq(2)
my_q.enq(4)
my_q.print()
print(my_q.peek_head()) 
print(my_q.peek_tail()) 
print("size =",my_q.qsize())

    


    
    
```

## Create the Node and Tree classes

```python
class Node:
    def __init__(self, value = None, left = None, right = None):
        self.value = value
        self.left  = left
        self.right = right
    
    def set_value(self,value):
        self.value = value
    
    def set_left_child(self, left):
        self.left = left
    
    def set_right_child(self, right):
        self.right = right
        
    def get_value(self):
        return self.value
    
    def get_left_child(self):
        return self.left
    
    def get_right_child(self):
        return self.right
    
    def has_value(self):
        if(self.value !=None):
            return True
        else:
            return False
    
    def has_left_child(self):
        if(self.left !=None):
            return True
        else:
            return False
    
    def has_right_child(self):
        if(self.right!=None):
            return True
        else:
            return False

class Tree:
    def __init__(self, root_node = None, root_node_value = None):
        if(root_node !=None):
            self.root = root_node
        elif(root_node_value !=None):
            self.root = Node(root_node_value)
        else:
            self.root = Node()
    
    def get_root(self):
        return self.root
    
    def print_details(self,current_node,visit_values,main_stack):
        print("current_node.value :",current_node.get_value())
        print("visit_values       : ",visit_values)
        print("main_stack-values  :"); main_stack.print_values()
        print("")
    
    def bfs(self,detail=0):
        current_node = self.root
        visit_order  = []
        visit_values = []
        
        bfs_q        = Queue()
        bfs_q.enq(current_node)       
        
        while(bfs_q.is_empty()==False):
            current_node = bfs_q.deq()
            visit_order.append(current_node)
            visit_values.append(current_node.get_value())
            if(current_node.has_left_child()):
                bfs_q.enq(current_node.get_left_child())
            if(current_node.has_right_child()):
                bfs_q.enq(current_node.get_right_child())
        
        if(detail ==1):
            print("final: visit_values",visit_values)

        return visit_values, visit_order
                
            
            

    
    
```

## Instantiate a Tree with values: i.e a Tree with Fruits

```python
node0 = Node("apple")
node1 = Node("banana")
node2 = Node("cherry")
node3 = Node("dates")

node0.set_left_child(node1)
node0.set_right_child(node2)

node1.set_left_child(node3)

print("node0:value=",node0.value,", left=" ,node0.left.value, ", right=" ,node0.right.value)
print("node1:value=",node1.value,", left=" ,node1.left.value, ", right=" ,node1.right)
print("node2:value=",node2.value,", left=" ,node2.left,       ", right=" ,node2.right)
print("node3:value=",node3.value,", left=" ,node3.left,       ", right=" ,node3.right)


my_tree0    = Tree(node0)
root_node   = my_tree0.get_root()
left_base   = root_node.get_left_child()
right_base  = root_node.get_right_child()
print("\n")
print("root_node  :value=",root_node.value ,", left=" ,root_node.left.value, ", right=" ,root_node.right.value)
print("left_base  :value=",left_base.value ,", left=" ,left_base.left.value, ", right=" ,left_base.right)
print("right_base :value=",right_base.value,", left=" ,right_base.left,      ", right=" ,right_base.right)

bfs_visit_list,_ = my_tree0.bfs(0)
print(bfs_visit_list)
```
