```python
from collections import deque

class Queue:
    def __init__(self,values = None):
        self.queue = deque()
        if(values):
            for i in values:
                self.queue.append(i)
    
    #Think of the deque as a list from left to right
    #I am going to enq on the right and deq on the left (this is the opposite of the udacity videos)
    def enq(self, value):
        self.queue.append(value)
    
    def deq(self):
        return self.queue.popleft()
    
    def peek_head(self):
        return self.queue[0]
    
    def peek_tail(self):
        return self.queue[-1]
    
    def qsize(self):
        return len(self.queue)
    
    def is_empty(self):
        if(len(self.queue)==0):
            return True
        else:
            return False
    
    def print(self):
        print(self.queue)
```

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.left  = None
        self.right = None
    
    def set_value(self,value):
        self.value =value
       
    def set_left_child(self, node=None):
        self.left = node

    def set_right_child(self,node=None):
        self.right = node
    
    def get_value(self):
        if(self.value !=None):
            return self.value
        else :
            return None 

    def get_left_child(self):
        if(self.left !=None):
            return self.left
        else :
            return None

    def get_right_child(self):
        if(self.right !=None):
            return self.right
        else :
            return None
        
    def has_left_child(self):
        if(self.left !=None):
            return True
        else :
            return False    

        
    def has_right_child(self):
        if(self.right !=None):
            return True
        else :
            return False
    
```

# class binary_search_tree:
## delete: method
In the delete method
#### When the current_node has no children or only 1 child  
The current node gets deleted/ replaced. In this case you need to keep track of the  
-> current_node  
-> current_node's parent  
-> current_node's child: so that when the node is getting deleted/repalced, the current_node's parent points to the current_node's child

#### When the current_node has 2 children
The current_node.value gets replaced with its inorder-predecessor value.  
NOTe: Only the current_node.value gets replaced, the current_node structure itself does not gets replaced . So there is no need to track the current_node's parent and current_node's child.  
Instead of the current_node,the inorder-predecessor gets deleted/replaced. Hence in this case need to keep track of the  
-> inorder-predecessor-node  
-> inorder-predecessor-node's parent  
-> inorder-predecessor-node's child: so that when the node is getting deleted/repalced, the inorder-predecessor-node's parent points to the inorder-predecessor-node's child  
FINER DETAILS: The inorder-predecessor can be the current_node's child : ip_flag ="child"  
The inorder-predecessor can be the current_node's grand-child, great grand-child etc : ip_flag ="grand-child"  

#### When current_node is root
For all 3 cases current_node has no children, only 1 child, has 2 children, when the current_node is root. Special attention needs to be given 

```python

class Binary_Search_Tree:
    def __init__(self,value = None, arr = False):
        if(value == None):
            self.root = None
        #Create a BST : the root only with 1 value
        elif(arr==False):
            self.root = value
        #Create a BST from an array
        else:
            self.root = Node(value[0])
            for idx in range(1, len(value)):
                #this is a very long process: see if it can be made efficient
                self.insert_by_loop(value[idx])
            
    
    def set_root(self, node):
        self.root = node
    
    def get_root(self):
        if (self.root != None): 
            return self.root
        else :
            return None 
        
    def bfs(self,detail=0):
        current_node = self.root
        visit_order  = []
        visit_values = []

        bfs_q        = Queue()
        bfs_q.enq(current_node)       

        while(bfs_q.is_empty()==False):
            current_node = bfs_q.deq()
            visit_order.append(current_node)
            visit_values.append(current_node.get_value())
            if(current_node.has_left_child()):
                bfs_q.enq(current_node.get_left_child())
            if(current_node.has_right_child()):
                bfs_q.enq(current_node.get_right_child())

        if(detail ==1):
            print("final: visit_values",visit_values)

        return visit_values, visit_order
    
    def search_by_loop(self, value):
        #==============BASE CASE: The root(No parent). =======================================
        #Since The root is the first node , it has no parent. 
        #(The while loop down below takes care of the rest of the nodes which all have a parent each)
        current_node = self.root
        #---------------------value==current_node.value---------------------
        if(value==current_node.value):
            search_flag = True
            parent_node = None
            lr_flag          = "root"
            return search_flag, current_node, parent_node, lr_flag
        #---------------------value<current_node.value---------------------
        elif(value<current_node.value):
            parent_node  = current_node
            if(parent_node.has_left_child()):
                current_node = parent_node.get_left_child()
                lr_flag           = "left"
                #now control will get transferred to the while loop
            else: 
                search_flag  = False
                current_node = None
                parent_node  = None
                lr_flag           = None
                return search_flag, current_node, parent_node,lr_flag
        #---------------------value>current_node.value---------------------
        else:
            parent_node  = current_node
            if(parent_node.has_right_child()):
                current_node = parent_node.get_right_child()
                lr_flag           = "right"
                #now control will get transferred to the while loop
            else: 
                search_flag  = False
                current_node = None
                parent_node  = None
                lr_flag           = None
                return search_flag, current_node, parent_node,lr_flag
            
        
        # =========== ALL OTHER NODES(have a parent each) ===========================
        #This while loop down takes care of the rest of the nodes which all have a parent each
        while(current_node !=None):
            #---------------------value==current_node.value---------------
            if(value==current_node.value):
                search_flag= True
                break
            #---------------------value<current_node.value----------------
            elif(value<current_node.value):
                parent_node  = current_node
                if(parent_node.has_left_child()):
                    current_node = parent_node.get_left_child()
                    lr_flag           = "left"
                else: 
                    search_flag  = False
                    current_node = None
                    parent_node  = None
                    lr_flag           = None
                    break
            #---------------------value>current_node.value----------------   
            else:
                parent_node  = current_node
                if(parent_node.has_right_child()):
                    current_node = parent_node.get_right_child()
                    lr_flag           = "right"
                else: 
                    search_flag  = False
                    current_node = None
                    parent_node  = None
                    lr_flag           = None
                    break
                    
        return search_flag, current_node, parent_node,lr_flag
    
    
    def insert_by_loop(self, value):
        current_node = self.root
        while(current_node !=None):
            if(value==current_node.value):
                return False
            elif(value<current_node.value):
                if(current_node.has_left_child()):
                    current_node = current_node.get_left_child()
                else: 
                    node_to_insert = Node(value)
                    current_node.set_left_child(node_to_insert)
                    return True
            else:
                if(current_node.has_right_child()):
                    current_node = current_node.get_right_child()
                else: 
                    node_to_insert = Node(value)
                    current_node.set_right_child(node_to_insert)
                    return True
        
        return False
    
    
    
    def delete(self,value):
        element_found, current_node, parent_node,lr_flag = self.search_by_loop(value)
        #Case 0: Element not found
        if(element_found = False):
            return element_found
        #Case 1: Element has no children
        elif(current_node.has_left_child() == False and current_node.has_right_child()==False):
            #current_node is root(Doesn't have a parent). Delete the root i.e. empty bst
            if(lr_flag=="root"): 
                current_node = None
            #current_node is the left_child: Set parent_node.left = None
            elif(lr_flag=="left"):
                parent_node.set_left_child()
            #current_node is the right_child: Set parent_node.right = None
            elif(lr_flag=="right"):
                parent_node.set_right_child()
            else:
                print("delete: Case1 ERROR")
                return      
                
        #Case 2: Element has only 1 child: has left-child
        elif(current_node.has_left_child() == True  and current_node.has_right_child()==False):
            child_node = current_node.get_left_child()
            #current_node is root(Doesn't not have a parent).Set child_node as the new root 
            if(lr_flag=="root"): 
                self.set_root(child_node)
            #current_node is the left_child: Set parent_node.left = child_node
            elif(lr_flag=="left"):
                parent_node.set_left_child(child_node)
            #current_node is the right_child: Set parent_node.right = child_node
            elif(lr_flag=="right"):
                parent_node.set_right_child(child_node)
            else:
                print("delete: Case2 ERROR")
                return
        
        #Case 3: Element has only 1 child: has right-child
        elif(current_node.has_left_child() == False  and current_node.has_right_child()==True):
            child_node = current_node.get_right_child()
            #current_node is root(Doesn't not have a parent).Set child_node as the new root 
            if(lr_flag=="root"): 
                self.set_root(child_node)
            #current_node is the left_child: Set parent_node.left = child_node
            elif(lr_flag=="left"):
                parent_node.set_left_child(child_node)
            #current_node is the right_child: Set parent_node.right = child_node
            elif(lr_flag=="right"):
                parent_node.set_right_child(child_node)
            else:
                print("delete: Case3 ERROR")
                return
        
        #Case 4: Element has 2 children: replace with inorder-predecessor
        else:
            #ip: inorder-predecessor
            #ipp:inorder-predecessor's parent
            #ipc:inorder-predecessor's child
            #ip_flag = "child" if inorder predecessor is the current node's child
            #ip_flag = "grand_child" in inorder predecesoor is current_node's grand_child, great_grand_child etc. 
            ip,ipp,ipc,ip_flag = self.find_inorder_predecessor(current_node)
            #the inorder-predecessor is the current_node's immediate child (current_node does not have grand_children)
            if(ip_flag == "child"):
                current_node.set_value(ip.get_value())
                current_node.set_left_child() #Set the current_node.left_child to null
            #inorder-predecessor is grandchild. but ip itself has no more children
            elif(ip_flag=="grand-child" and ipc==None):
                current_node.set_value(ip.get_value())
                ipp.set_right_child() #Set the ipp.right_child to null
            #inorder-predecessor is grandchild. ip itself has no left children
            elif(ip_flag=="grand-child" and ipc==None):
                current_node.set_value(ip.get_value())
                ipp.set_left_child(ipc) #Set the ipp.right_child to null                
```
