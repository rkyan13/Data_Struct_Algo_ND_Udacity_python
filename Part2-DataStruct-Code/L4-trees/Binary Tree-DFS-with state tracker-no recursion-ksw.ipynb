## Create the data structures necessary for DFS

```python
class Stack:
    def __init__(self):
        self.list = []
    
    def push(self,item):
        self.list.append(item)
    
    def pop(self):
        if(self.size()!=0):
            return self.list.pop()
        else:
            return None        
    
    def peek(self):
        if(self.size()!=0):
            return self.list[-1]
        else:
            return None
    
    def is_empty(self):
        if(len(self.list)==0):
            return True
        else:
            return False
    
    def size(self):
        return len(self.list)
    
    def print(self):
        for i in range(0,len(self.list)):
            idx=len(self.list)-i-1
            print(self.list[idx])
    
            
    def print_values(self):
        for i in range(0,len(self.list)):
            print(self.list[i].node.value, end =" ")


#Test the stack
dfs_stack = Stack()
dfs_stack.push(1)
dfs_stack.push(2)
dfs_stack.push(3)
dfs_stack.push(4)
dfs_stack.push(5)
dfs_stack.print()
```

## Create the Node and Tree classes

```python
class Node:
    def __init__(self, value = None, left = None, right = None):
        self.value = value
        self.left  = left
        self.right = right
    
    def set_value(self,value):
        self.value = value
    
    def set_left_child(self, left):
        self.left = left
    
    def set_right_child(self, right):
        self.right = right
        
    def get_value(self):
        return self.value
    
    def get_left_child(self):
        return self.left
    
    def get_right_child(self):
        return self.right
    
    def has_value(self):
        if(self.value !=None):
            return True
        else:
            return False
    
    def has_left_child(self):
        if(self.left !=None):
            return True
        else:
            return False
    
    def has_right_child(self):
        if(self.right!=None):
            return True
        else:
            return False

class State:
    def __init__(self, node):
        self.node = node
        
        self.visited = False
        
        if(self.node.has_left_child()):
            self.left_child_visited = False
        else:
            self.left_child_visited = True
        
        if(self.node.has_right_child()):
            self.right_child_visited = False
        else:
            self.right_child_visited = True
        
    def get_visited(self):
        return self.visited
                   
    def get_left_child_visited(self):
        return self.left_child_visited 
    
    def get_right_child_visited(self):
        return self.right_child_visited 
    
    def set_left_child_visited(self,status):
        self.left_child_visited = status
    
    def set_right_child_visited(self, status):
        self.right_child_visited = status
    
    def set_visited(self,status):
        self.visited = status
    
    
```

```python
class Tree:
    def __init__(self, root_node = None, root_node_value = None):
        if(root_node !=None):
            self.root = root_node
        elif(root_node_value !=None):
            self.root = Node(root_node_value)
        else:
            self.root = Node()
    
    def get_root(self):
        return self.root
    
    def print_details(self,current_node,visit_values,main_stack):
        print("current_node.value :",current_node.get_value())
        print("visit_values       : ",visit_values)
        print("main_stack-values  :"); main_stack.print_values()
        print("")

    def dfs_preorder(self,detail=0):    
        visit_order    = [] #visit_order contains the pointer to the state of the nodes
        visit_values   = [] #visit_values contains the actual values of the nodes
        main_stack     = Stack()


        #The very first initial entry
        current_node  = self.get_root()
        visit_values.append(current_node.get_value())
        current_state = State(current_node)
        visit_order.append(current_state)        
        main_stack.push(current_state)
        iteration =0
        if(detail ==1):
                print("\n----------Start state: while loop:----------------")
                self.print_details(current_node,visit_values,main_stack)

                
        while(current_node!=None):
            iteration = iteration +1
            if(detail ==1):
                print("\n----------BEGIN:iteration#:",iteration," ----------------")
                print("Begin of iteration status:::::::")
                self.print_details(current_node,visit_values,main_stack)
                
            #-----------------BEGIN: Traverse down the Left Subtree--------------------------------------------------------------------
            #if the left child has not yet been traversed...........
            if(current_state.get_left_child_visited()==False):
                if(detail ==1): print("Case 1a-Traversing.. left subtree !")      
                #i)   change the left-child-visited status of current node to True
                current_state.set_left_child_visited(True)
                #ii)  make the new-left-node the current_node 
                #iii) push the current_state (i.e. new-left-node) to the visit order and main_stack
                current_node  = current_node.get_left_child()
                visit_values.append(current_node.get_value())
                current_state = State(current_node)
                visit_order.append(current_state)                
                main_stack.push(current_state)
                if(detail ==1):  
                    print("EOF iteration status:::::::")
                    self.print_details(current_node,visit_values,main_stack)
                    print("----------END:iteration#:",iteration," ----------------")
                #Continue to keep checking the left subtrees until you encounter a leaf
                continue 
            #else: the left subtree has already been traversed...........
            else:
                if(detail ==1): print("Case 1b-left subtree already traversed/(does not exist) !: no mods")                 
            #-----------------EOF: Traverse down the Left Subtree--------------------------------------------------------------------


            #-----------------BEGIN: Traverse down the Right Subtree----------------------------------------------------------------------
            #Traverse the Immediate right child: But don't keep recursively heading down the right side. 
            #After the right child, explore left grandchildren of the right child
            #if the right subtree has not yet been traversed...........
            if(current_state.get_right_child_visited()==False):
                if(detail ==1): print("Case 2a-Traversing.. right subtree !")
                #if the right child has not yet visited...........
                #i) change the right-child-visited status of current node to True
                current_state.set_right_child_visited(True)
                #ii)  make the new-right-node the current_node 
                #iii) push the current_state (i.e. new-left-node) to the visit order and main_stack
                current_node  = current_node.get_right_child()
                visit_values.append(current_node.get_value())
                current_state = State(current_node)
                visit_order.append(current_state)                
                main_stack.push(current_state)


            #if the right subtree has already been traversed..........
            #This means the end of the right subtree (and hence the end of traversal of both left+right subtrees)
            #Now we can travel up one level and start searching the parent-node's siblings' subtrees
            #pop the relevant main_stack
            else:
                if(detail ==1): print("Case 2b-right subtree already traversed(hence left and right tree traversed.: ")
                #Pop the node of the main stack : indicates both left-subtree and right-subtree has been traversed
                main_stack.pop()
                #Update the current_node
                current_state = main_stack.peek()
                if(current_state !=None):
                    current_node  = current_state.node        
                else:
                    current_node  = None                
            #-----------------END: Traverse down the Right Subtree---------------------------------------------
            
            if(detail ==1 and current_node!=None):                
                self.print_details(current_node,visit_values,main_stack)
                print("----------End:iteration#:",iteration," ----------------")

        #EOF: while(current_node!=None):---------------------------------------------------------------
        
        if(detail ==1):
            print("final: visit_values",visit_values)

        return visit_values, visit_order
    
    
    
    def dfs_inorder(self,detail=0):    
        visit_order    = [] #visit_order contains the pointer to the state of the nodes
        visit_values   = [] #visit_values contains the actual values of the nodes
        main_stack     = Stack()


        #The very first initial entry
        current_node  = self.get_root()
        current_state = State(current_node)   
        main_stack.push(current_state)
        iteration =0
        if(detail ==1):
                print("\n----------Start state: while loop:----------------")
                self.print_details(current_node,visit_values,main_stack)

                
        while(current_node!=None):
            iteration = iteration +1
            if(detail ==1):
                print("\n----------BEGIN:iteration#:",iteration," ----------------")
                print("Begin of iteration status:::::::")
                self.print_details(current_node,visit_values,main_stack)
                
            #-----------------BEGIN: Traverse down the Left Subtree--------------------------------------------------------------------
            if(current_state.get_left_child_visited()==False):
                if(detail ==1): print("Case 1a-Traversing.. left subtree !")
                #if the left child has not yet visited...........
                #i)   change the left-child-visited status of current node to True
                current_state.set_left_child_visited(True)
                #ii)  make the new-left-node the current_node 
                #iii) push the current_state (i.e. new-left-node) to the visit order and main_stack
                current_node  = current_node.get_left_child()
                current_state = State(current_node)             
                main_stack.push(current_state)
                if(detail ==1):  
                    print("EOF iteration status:::::::")
                    self.print_details(current_node,visit_values,main_stack)
                    print("----------END:iteration#:",iteration," ----------------")
                #Continue to keep checking the left subtrees until you encounter a leaf
                continue 
            #else: the left subtree has already been traversed...........
            else:
                if(detail ==1): print("Case 1b-left subtree already traversed/(does not exist).might/might-not mod the Visit order !: ") 
                if(current_state.get_visited()==False):
                    current_state.set_visited(True)
                    visit_values.append(current_node.get_value())
                    visit_order.append(current_state) 
                if(detail ==1):  self.print_details(current_node,visit_values,main_stack)
            #-----------------EOF: Traverse down the Left Subtree--------------------------------------------------------------------


            #-----------------BEGIN: Traverse down the Right Subtree----------------------------------------------------------------------
            #Traverse the Immediate right child: But don't keep recursively heading down the right side. 
            #After the right child, explore left grandchildren of the right child
            #if the right subtree has not yet been traversed...........
            if(current_state.get_right_child_visited()==False):
                if(detail ==1): print("Case 2a-Traversing.. right subtree !Modifying visit order")
                #if the right child has not yet visited...........
                #i) change the right-child-visited status of current node to True
                current_state.set_right_child_visited(True)
                #ii)  make the new-right-node the current_node 
                #iii) push the current_state (i.e. new-left-node) to the visit order and main_stack
                current_node  = current_node.get_right_child()
                current_state = State(current_node)            
                main_stack.push(current_state)


            #if the right subtree has already been traversed..........
            #This means the end of the right subtree (and hence the end of traversal of both left+right subtrees)
            #Now we can travel up one level and start searching the parent-node's siblings' subtrees
            #pop the relevant main_stack
            else:
                if(detail ==1): print("Case 2b-right subtree already traversed(hence left and right tree traversed.: ")
                #Pop the node of the main stack : indicates both left-subtree and right-subtree has been traversed
                main_stack.pop()
                #Update the current_node
                current_state = main_stack.peek()
                if(current_state !=None):
                    current_node  = current_state.node        
                else:
                    current_node  = None                
            #-----------------END: Traverse down the Right Subtree---------------------------------------------
            
            if(detail ==1 and current_node!=None):                
                self.print_details(current_node,visit_values,main_stack)
                print("----------End:iteration#:",iteration," ----------------")

        #EOF: while(current_node!=None):---------------------------------------------------------------
        
        if(detail ==1):
            print("final: visit_values",visit_values)

        return visit_values, visit_order
    
    
    

    

    def dfs_postorder(self,detail=0):    
            visit_order    = [] #visit_order contains the pointer to the state of the nodes
            visit_values   = [] #visit_values contains the actual values of the nodes
            main_stack     = Stack()


            #The very first initial entry
            current_node  = self.get_root()
            current_state = State(current_node)   
            main_stack.push(current_state)
            iteration =0
            if(detail ==1):
                    print("\n----------Start state: while loop:----------------")
                    self.print_details(current_node,visit_values,main_stack)


            while(current_node!=None):
                iteration = iteration +1
                if(detail ==1):
                    print("\n----------BEGIN:iteration#:",iteration," ----------------")
                    print("Begin of iteration status:::::::")
                    self.print_details(current_node,visit_values,main_stack)

                #-----------------BEGIN: Traverse down the Left Subtree--------------------------------------------------------------------
                if(current_state.get_left_child_visited()==False):
                    if(detail ==1): print("Case 1a-Traversing.. left subtree !")
                    #if the left child has not yet visited...........
                    #i)   change the left-child-visited status of current node to True
                    current_state.set_left_child_visited(True)
                    #ii)  make the new-left-node the current_node 
                    #iii) push the current_state (i.e. new-left-node) to the visit order and main_stack
                    current_node  = current_node.get_left_child()
                    current_state = State(current_node)             
                    main_stack.push(current_state)
                    if(detail ==1):  
                        print("EOF iteration status:::::::")
                        self.print_details(current_node,visit_values,main_stack)
                        print("----------END:iteration#:",iteration," ----------------")
                    #Continue to keep checking the left subtrees until you encounter a leaf
                    continue 
                #else: the left subtree has already been traversed...........
                else:
                    if(detail ==1): print("Case 1b-left subtree already traversed/(does not exist).might/might-not mod the Visit order !: ") 

                #-----------------EOF: Traverse down the Left Subtree--------------------------------------------------------------------


                #-----------------BEGIN: Traverse down the Right Subtree----------------------------------------------------------------------
                #Traverse the Immediate right child: But don't keep recursively heading down the right side. 
                #After the right child, explore left grandchildren of the right child
                #if the right subtree has not yet been traversed...........
                if(current_state.get_right_child_visited()==False):
                    if(detail ==1): print("Case 2a-Traversing.. right subtree !Modifying visit order")
                    #if the right child has not yet visited...........
                    #i) change the right-child-visited status of current node to True
                    current_state.set_right_child_visited(True)
                    #ii)  make the new-right-node the current_node 
                    #iii) push the current_state (i.e. new-left-node) to the visit order and main_stack
                    current_node  = current_node.get_right_child()
                    current_state = State(current_node)            
                    main_stack.push(current_state)


                #if the right subtree has already been traversed..........
                #This means the end of the right subtree (and hence the end of traversal of both left+right subtrees)
                #Now we can travel up one level and start searching the parent-node's siblings' subtrees
                #pop the relevant main_stack
                else:
                    if(detail ==1): print("Case 2b-right subtree already traversed(hence left and right tree traversed, mod visit order ")
                    if(current_state.get_visited()==False):
                        current_state.set_visited(True)
                        visit_values.append(current_node.get_value())
                        visit_order.append(current_state)                  
                    
                    #Pop the node of the main stack : indicates both left-subtree and right-subtree has been traversed
                    main_stack.pop()
                    #Update the current_node
                    current_state = main_stack.peek()
                    if(current_state !=None):
                        current_node  = current_state.node        
                    else:
                        current_node  = None                
                #-----------------END: Traverse down the Right Subtree---------------------------------------------

                if(detail ==1 and current_node!=None):                
                    self.print_details(current_node,visit_values,main_stack)
                    print("----------End:iteration#:",iteration," ----------------")

            #EOF: while(current_node!=None):---------------------------------------------------------------

            if(detail ==1):
                print("final: visit_values",visit_values)

            return visit_values, visit_order

    


    
    
```
