## Create the Node and Tree classes

```python
class Node:
    def __init__(self, value = None, left = None, right = None):
        self.value = value
        self.left  = left
        self.right = right
    
    def set_value(self,value):
        self.value = value
    
    def set_left_child(self, left):
        self.left = left
    
    def set_right_child(self, right):
        self.right = right
        
    def get_value(self):
        return self.value
    
    def get_left_child(self):
        return self.left
    
    def get_right_child(self):
        return self.right
    
    def has_value(self):
        if(self.value !=None):
            return True
        else:
            return False
    
    def has_left_child(self):
        if(self.left !=None):
            return True
        else:
            return False
    
    def has_right_child(self):
        if(self.right!=None):
            return True
        else:
            return False
        

class Tree:
    def __init__(self, root_node = None, root_node_value = None):
        if(root_node !=None):
            self.root = root_node
        elif(root_node_value !=None):
            self.root = Node(root_node_value)
        else:
            self.root = Node()
    
    def get_root(self):
        return self.root

    def dfs_preorder(self,detail=0):    
        visit_order  = [] #visit_order contains the pointer to the nodes
        visit_values = [] #visit_values contains the actual values of the nodes
        main_stack   = Stack()
        left_parents   = Stack()
        right_parents  = Stack()

        #The very first initial entry
        current_node = self.get_root()
        visit_order.append(current_node)
        visit_values.append(current_node.get_value())
        main_stack.push(current_node)
        iteration =0
        if(detail ==1):
                print("\n----------Start state: while loop:----------------")
                print("current_node.value :",current_node.get_value())
                print("visit_values       : ",visit_values)
                print("main_s9tack-values  :"); main_stack.print_values()
                print("\nleft_parents-values  :"); left_parents.print_values()
                print("\nright_parents-values :"); right_parents.print_values()
                print("")
        while(current_node!=None):
            iteration = iteration +1
            if(detail ==1):
                print("\n----------BEGIN:iteration#:",iteration," ----------------")
                print("Begin of iteration status:::::::")
                print("current_node.value :",current_node.get_value())
                print("visit_values       : ",visit_values)
                print("main_stack-values  :"); main_stack.print_values()
                print("\nleft_parents-values  :"); left_parents.print_values()
                print("\nright_parents-values :"); right_parents.print_values()
                print("")
            #-----------------If left_child_exists-----------------------------------------------------------------------
            #Traverse down the Left Subtree
            if(current_node.has_left_child()):
                #if the left subtree has not yet been traversed...........
                #i) push the current node to the left_parents
                #ii) make the new-left-node the current_node 
                #iii) push the current_node (i.e. new-left-node) to the visit order and main_stack
                if(left_parents.peek()!=current_node):
                    if(detail ==1):
                        print("Case 1a-Left-child exists: left subtree traversing..: ")
                        print("modifies left_parents, current_node, visit_order,main_stack")
                    left_parents.push(current_node)            
                    current_node = current_node.get_left_child()
                    visit_order.append(current_node)
                    visit_values.append(current_node.get_value())
                    main_stack.push(current_node)
                    if(detail ==1):  
                        print("EOF iteration status:::::::")
                        print("current_node.value :",current_node.get_value())
                        print("visit_values       : ",visit_values)
                        print("main_stack-values  :"); main_stack.print_values()
                        print("\nleft_parents-values  :"); left_parents.print_values()
                        print("\nright_parents-values :"); right_parents.print_values()
                        print("")
                        print("----------END:iteration#:",iteration," ----------------")
                    #Continue to keep checking the left subtrees until you encounter a leaf
                    continue 
                #else: the left subtree has already been traversed...........
                else:
                    if(detail ==1):
                        print("Case 1b-Left-child exists: left subtree already traversed !: no mods")
            #-----------------else left_child does not exist-------------------------------------
            else:
                if(detail ==1):
                    print("Case 1c-Left-child does not exist! : no mods")                       
            #-----------------EOF: If left_child_exists--------------------------------------------------------------------


            #-----------------If right_child_exists-------------------------------------------------------------------------
            #Traverse the Immediate right child: But don't keep recrusively heading down the right side. 
            #After the right child, explore left grandchildren of the right child
            if(current_node.has_right_child()):
                #if the right subtree has not yet been traversed...........
                #i) push the current node to the right_parents
                #ii) make the new-right-node the current_node 
                #iii) push the current_node (i.e. new-right-node) to the visit order and main_stack
                if(right_parents.peek()!=current_node):
                    if(detail ==1):
                        print("Case 2a-right-child exists: right subtree traversing..: ")
                        print("modifies right_parents, current_node, visit_order,main_stack")
                    right_parents.push(current_node)
                    current_node = current_node.get_right_child()
                    visit_order.append(current_node)
                    visit_values.append(current_node.get_value())
                    main_stack.push(current_node)

                #if the right subtree has already been traversed..........
                #This too means the end of the right subtree (and hence the end of traversal of both left+right subtrees)
                #Now we can travel up one level and start searching the parent-node's siblings' subtrees
                #pop the relevant main_stack, left_parents, right_parents
                else:
                    if(detail ==1):
                        print("Case 2b-right-child exists: right subtree already traversed: ")
                        print("modifies left_parents,right_parents, main_stack,current_node")
                    #if True it means there existed a left-subtree(left-child+its descendants) that has been traversed. Hence pop it off the left_parents
                    #if False it means that there was no left-subtree. so no need to pop the left_parents
                    if(left_parents.peek()==current_node):
                        left_parents.pop()
                    #if True it means there existed a right-subtree(right-child+its descendants) that has been traversed. Hence pop it off the right_parents
                    #if False it means that there was no right-subtree.so no need to pop the right_parents
                    if(right_parents.peek()==current_node):
                        right_parents.pop()
                    #Pop the node of the main stack : indicates both left-subtree and right-subtree has been traversed
                    main_stack.pop()
                    #Update the current_node
                    current_node = main_stack.peek()

            #-------else:right_child does not exist -------------------------------------------------------------------------
            #This too means the end of the right subtree (and hence the end of traversal of both left+right subtrees)
            #Now we can travel up one level and start searching the parent-node's siblings' subtrees
            #pop the relevant main_stack, left_parents, right_parents
            else:
                if(detail ==1):
                    print("Case 2c-right-child does not exist!")   
                    print("modifies left_parents,right_parents, main_stack,current_node")
                #if True it means there existed a left-subtree(left-child+its descendants) that has been traversed. Hence pop it off the left_parents
                #if False it means that there was no left-subtree. so no need to pop the left_parents
                if(left_parents.peek()==current_node):
                    left_parents.pop()
                #if True it means there existed a right-subtree(right-child+its descendants) that has been traversed. Hence pop it off the right_parents
                #if False it means that there was no right-subtree.so no need to pop the right_parents
                if(right_parents.peek()==current_node):
                    right_parents.pop()
                #Pop the node of the main stack : indicates both left-subtree and right-subtree has been traversed
                main_stack.pop()
                #Update the current_node
                current_node = main_stack.peek()
            #----------------EOF:if-else right_child_exists--------------------------------------------------------------------
            
            if(detail ==1 and current_node!=None):                
                print("current_node.value :",current_node.get_value())
                print("visit_values       : ",visit_values)
                print("main_stack-values  :"); main_stack.print_values()
                print("\nleft_parents-values  :"); left_parents.print_values()
                print("\nright_parents-values :"); right_parents.print_values()
                print("")
                print("----------End:iteration#:",iteration," ----------------")

        #EOF: while(current_node!=None):---------------------------------------------------------------
        if(detail ==1):
            print("final: visit_values",visit_values)

        return visit_values, visit_order

    
    def dfs_inorder(self,detail=0):    
        visit_order  = [] #visit_order contains the pointer to the nodes
        visit_values = [] #visit_values contains the actual values of the nodes
        main_stack   = Stack()
        left_parents   = Stack()
        right_parents  = Stack()
        right_children = Stack()

        #The very first initial entry
        current_node = self.get_root()
        main_stack.push(current_node)
        iteration =0
        if(detail ==1):
                print("\n----------Start state: while loop:----------------")
                print("current_node.value :",current_node.get_value())
                print("visit_values       : ",visit_values)
                print("main_stack-values  :"); main_stack.print_values()
                print("\nleft_parents-values  :"); left_parents.print_values()
                print("\nright_parents-values :"); right_parents.print_values()
                print("\nright_children-values :"); right_children.print_values()
                print("")
        while(current_node!=None):
            iteration = iteration +1
            if(detail ==1):
                print("\n----------BEGIN:iteration#:",iteration," ----------------")
                print("Begin of iteration status:::::::")
                print("current_node.value :",current_node.get_value())
                print("visit_values       : ",visit_values)
                print("main_stack-values  :"); main_stack.print_values()
                print("\nleft_parents-values  :"); left_parents.print_values()
                print("\nright_parents-values :"); right_parents.print_values()
                print("\nright_children-values :"); right_children.print_values()
                print("")
            #-----------------If left_child_exists-----------------------------------------------------------------------
            #Traverse down the Left Subtree
            if(current_node.has_left_child()):
                #if the left subtree has not yet been traversed...........
                #i) push the current node to the left_parents
                #ii) make the new-left-node the current_node 
                #iii) push the current_node (i.e. new-left-node) to the visit order and main_stack
                if(left_parents.peek()!=current_node):
                    if(detail ==1):
                        print("Case 1a-Left-child exists: left subtree traversing..: ")
                        print("modifies left_parents, current_node, main_stack")
                    left_parents.push(current_node)            
                    current_node = current_node.get_left_child()
                    main_stack.push(current_node)
                    if(detail ==1):  
                        print("EOF iteration status:::::::")
                        print("current_node.value :",current_node.get_value())
                        print("visit_values       : ",visit_values)
                        print("main_stack-values  :"); main_stack.print_values()
                        print("\nleft_parents-values  :"); left_parents.print_values()
                        print("\nright_parents-values :"); right_parents.print_values()
                        print("\nright_children-values :"); right_children.print_values()
                        print("")
                        print("----------END:iteration#:",iteration," ----------------")
                    #Continue to keep checking the left subtrees until you encounter a leaf
                    continue 
                #else: the left subtree has already been traversed...........
                else:
                    if(len(visit_values)==0 or current_node.has_right_child() == False or current_node.get_right_child()!=right_children.peek()):
                        visit_order.append(current_node)
                        visit_values.append(current_node.get_value())
                        if(detail ==1):
                            print("Case 1b-Left-child exists: left subtree already traversed !mods visit_order")
                    else:
                        right_children.pop()
                        if(detail ==1):
                            print("Case 1b-Left-child exists: left subtree already traversed !revisiting parent-pop-right-children-NO-MODS")
            #-----------------else left_child does not exist-------------------------------------
            else:
                if(len(visit_values)==0 or current_node.has_right_child() == False or current_node.get_right_child()!=right_children.peek()):
                    visit_order.append(current_node)
                    visit_values.append(current_node.get_value())
                    if(detail ==1):
                        print("Case 1c-Left-child does not exist! : mods visit_order")  
                else:
                    right_children.pop()
                    if(detail ==1):
                        print("Case 1c-Left-child does not exist!, revisiting parent-pop right-children but NO-other MODS")  
            #-----------------EOF: If left_child_exists--------------------------------------------------------------------


            #-----------------If right_child_exists-------------------------------------------------------------------------
            #Traverse the Immediate right child: But don't keep recrusively heading down the right side. 
            #After the right child, explore left grandchildren of the right child
            if(current_node.has_right_child()):
                #if the right subtree has not yet been traversed...........
                #i) push the current node to the right_parents
                #ii) make the new-right-node the current_node 
                #iii) push the current_node (i.e. new-right-node) to the visit order and main_stack
                if(right_parents.peek()!=current_node):
                    if(detail ==1):
                        print("Case 2a-right-child exists: right subtree traversing..: ")
                        print("modifies right_parents, current_node, main_stack, right_children")
                    right_parents.push(current_node)
                    current_node = current_node.get_right_child()
                    #visit_order.append(current_node)
                    #visit_values.append(current_node.get_value())
                    main_stack.push(current_node)
                    right_children.push(current_node)

                #if the right subtree has already been traversed..........
                #This too means the end of the right subtree (and hence the end of traversal of both left+right subtrees)
                #Now we can travel up one level and start searching the parent-node's siblings' subtrees
                #pop the relevant main_stack, left_parents, right_parents
                else:
                    if(detail ==1):
                        print("Case 2b-right-child exists: right subtree already traversed: ")
                        print("modifies left_parents,right_parents, main_stack,current_node")
                    #if True it means there existed a left-subtree(left-child+its descendants) that has been traversed. Hence pop it off the left_parents
                    #if False it means that there was no left-subtree. so no need to pop the left_parents
                    if(left_parents.peek()==current_node):
                        left_parents.pop()
                    #if True it means there existed a right-subtree(right-child+its descendants) that has been traversed. Hence pop it off the right_parents
                    #if False it means that there was no right-subtree.so no need to pop the right_parents
                    if(right_parents.peek()==current_node):
                        right_parents.pop()
                    #Pop the node of the main stack : indicates both left-subtree and right-subtree has been traversed
                    main_stack.pop()
                    #Update the current_node
                    current_node = main_stack.peek()

            #-------else:right_child does not exist -------------------------------------------------------------------------
            #This too means the end of the right subtree (and hence the end of traversal of both left+right subtrees)
            #Now we can travel up one level and start searching the parent-node's siblings' subtrees
            #pop the relevant main_stack, left_parents, right_parents
            else:
                if(detail ==1):
                    print("Case 2c-right-child does not exist!")   
                    print("modifies left_parents,right_parents, main_stack,current_node")
                #if True it means there existed a left-subtree(left-child+its descendants) that has been traversed. Hence pop it off the left_parents
                #if False it means that there was no left-subtree. so no need to pop the left_parents
                if(left_parents.peek()==current_node):
                    left_parents.pop()
                #if True it means there existed a right-subtree(right-child+its descendants) that has been traversed. Hence pop it off the right_parents
                #if False it means that there was no right-subtree.so no need to pop the right_parents
                if(right_parents.peek()==current_node):
                    right_parents.pop()
                #Pop the node of the main stack : indicates both left-subtree and right-subtree has been traversed
                main_stack.pop()
                #Update the current_node
                current_node = main_stack.peek()
            #----------------EOF:if-else right_child_exists--------------------------------------------------------------------
            
            if(detail ==1 and current_node!=None):                
                print("current_node.value :",current_node.get_value())
                print("visit_values       : ",visit_values)
                print("main_stack-values  :"); main_stack.print_values()
                print("\nleft_parents-values  :"); left_parents.print_values()
                print("\nright_parents-values :"); right_parents.print_values()
                print("\nright_children-values :"); right_children.print_values()
                print("")
                print("----------End:iteration#:",iteration," ----------------")

        #EOF: while(current_node!=None):---------------------------------------------------------------
        if(detail ==1):
            print("final: visit_values",visit_values)

        return visit_values, visit_order

 
```

## Create the data structures necessary for DFS

```python
class Stack:
    def __init__(self):
        self.list = []
    
    def push(self,item):
        self.list.append(item)
    
    def pop(self):
        if(self.size()!=0):
            return self.list.pop()
        else:
            return None        
    
    def peek(self):
        if(self.size()!=0):
            return self.list[-1]
        else:
            return None
    
    def is_empty(self):
        if(len(self.list)==0):
            return True
        else:
            return False
    
    def size(self):
        return len(self.list)
    
    def print(self):
        for i in range(0,len(self.list)):
            idx=len(self.list)-i-1
            print(self.list[idx])
    
    def print(self):
        for i in range(0,len(self.list)):
            idx=len(self.list)-i-1
            print(self.list[idx])
            
    def print_values(self):
        for i in range(0,len(self.list)):
            print(self.list[i].value, end =" ")


#Test the stack
dfs_stack = Stack()
dfs_stack.push(1)
dfs_stack.push(2)
dfs_stack.push(3)
dfs_stack.push(4)
dfs_stack.push(5)
dfs_stack.print()
```
